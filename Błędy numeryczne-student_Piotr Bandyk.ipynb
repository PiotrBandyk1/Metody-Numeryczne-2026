## Błąd względny i bezwzględny

***Zadanie 1.*** 
Liczbę $\textrm{e}$ możemy zdefiniować jako $\sum\limits_{n=0}^{\infty} \frac{1}{n!}$. Oblicz błąd względny i bezwzględny aproksymacji liczby $\textrm{e}$ poprzez $\textrm{e}^{*}$ w przypadku gdy:  
* $\textrm{e}_1^{*}=\sum\limits_{n=0}^{5} \frac{1}{n!}$    

* $\textrm{e}_2^{*}=\sum\limits_{n=0}^{10} \frac{1}{n!}$  

* $\textrm{e}_3^{*}=\sum\limits_{n=0}^{20} \frac{1}{n!}$  

```python
import math
import pandas as pd
from decimal import Decimal, getcontext
from IPython.display import display

getcontext().prec = 50

e_exact = Decimal(1).exp()
6

n_values = [5, 10, 20]

results = []

for n in n_values:

    e_approx = sum(Decimal(1) / Decimal(math.factorial(i)) for i in range(n + 1))

    abs_error = abs(e_exact - e_approx)

    rel_error = abs_error / e_exact

    results.append({
        "n": n,
        "Przybliżenie (e*)": float(e_approx), 
        "Błąd bezwzględny": f"{abs_error:.4E}", 
        "Błąd względny": f"{rel_error:.4E}"
    })

df = pd.DataFrame(results)
df.set_index("n", inplace=True)

display(df)
```

```python
print("1. Tak. Sprawdzamy to, obliczając wyznacznik – jeśli wynosi dokładnie 0, macierz jest osobliwa")
print("Analitycznie wyznacznik wynosi idealne 0, Numerycnie wynik nie wyniesie 0, lecz będzie mikroskopijną liczbą")

import numpy as np

A = np.array([
    [np.sqrt(2), 1/7],
    [np.pi * np.sqrt(2), np.pi / 7]
])

det_num = np.linalg.det(A)
print(f"Wyznacznik (np.linalg.det): {det_num}")
print(f"Czy wyznacznik to równe 0.0? {det_num == 0.0}\n")

try:
    A_odwrotna = np.linalg.inv(A)
    print("Udało się odwrócić macierz! Oto wynik (wielkie, 'śmieciowe' liczby):")
    print(A_odwrotna)
except np.linalg.LinAlgError as e:
    print(f"Nie udało się odwrócić macierzy. Błąd: {e}")
```
